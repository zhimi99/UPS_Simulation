```python
#importar las librerias necesarias
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#Construccion de un grafico basico
x = [0,1,2,3,4, 5]
y = [0,2,4,6,8, 10]
plt.figure(figsize=(7,5), dpi=50)#dpi tamaño del grafico y el numero de pixeles por pulgada.

# Graficos en forma de puntos 
plt.plot(x,y,'bo')
plt.show()

#Graficar de linea. intercortada
plt.plot(x,y, color='blue', linewidth=2, linestyle='--', label='Linea')
plt.show()
```

```python
# Generar un rango de puntos para ello vamos a utilizar np.arange(inicio, fin, paso)
x2 = np.arange(0,4.5,0.5)
y2 = np.arange(0,4.5,0.5)

plt.figure(figsize=(8,5))
plt.title('Grafico', fontdict={'fontname':'Comic Sans MS', 'fontsize': 20})
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

#Graficar
plt.plot(x2,y2,'b--')
#Segunda linea con puntos y elevado al cuadrado el arreglo x2
plt.plot(x2,x2**2,'ro')
# Asignar Escala de eje en X
plt.xticks([0,1,2,3,4])
#Agregar la leyenda al grafico
plt.legend()
#Guardar la imagen en un archivo
plt.savefig('grafico.png',dpi=300)
plt.show()
```

```python
# Graficar funciones matematicas
x3 = np.arange(0, 4*np.pi,0.1)
y3 = np.sin(x3)
plt.figure(figsize=(8,5))
plt.title('Grafico de SENO', fontdict={'fontname':'Comic Sans MS', 'fontsize': 20})
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.plot(x3,y3)
plt.show()
```

```python
#Grafico de Barras
etiquetas = ['A', 'B', 'C']
valores = np.arange(1,4,1)
plt.figure(figsize=(5,3))
barras = plt.bar(etiquetas, valores)
#Patrones dentro de las barras
patrones = ['/', 'o', '.']
for bar in barras:
    bar.set_hatch(patrones.pop(0))

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Trabajar con otro tipo de datos Fifa
fifa = pd.read_csv('fifa.csv')
fifa.head()
```

```python
## Pintado recto

bins = np.arange(40,110,10)

# PLT.HIST (FIFA.OBERTULT, BINS = BINS, COLOR = '# A87632') # El color aquí se puede seleccionar en el selector de color de búsqueda del navegador, luego copiar el código HEX está bien
plt.hist(fifa.overall_rating,bins=bins,color='black')  # Negro

plt.xticks(bins)
plt.yticks(np.arange(0,11000,1000))


plt.xlabel("Skill levels")
plt.ylabel("Number of Players")

plt.title("Distribution of Player Skills in FIFA 2018")

plt.show()
```

```python
fifa.weight_kgs.head()
```

```python
# Retire este símbolo de libra y conviértase en un número de punto flotante
fifa.weight_kgs = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.weight_kgs]
fifa.weight_kgs[0]
```

```python
light = fifa.loc[fifa.weight_kgs<56].count()[0]
print(light)
light_medium = fifa.loc[(fifa.weight_kgs>=56) & (fifa.weight_kgs<68)].count()[0]
print(light_medium)
medium = fifa.loc[(fifa.weight_kgs>=68) & (fifa.weight_kgs<79)].count()[0]
print(medium)
medium_heavy = fifa.loc[(fifa.weight_kgs>=79) & (fifa.weight_kgs<90)].count()[0]
print(medium_heavy)
heavy = fifa.loc[fifa.weight_kgs>=90].count()[0]
print(heavy)

plt.figure(figsize=(5,5),dpi=100)

#plt.style.use('default')
plt.style.use('ggplot')  # Directamente con algunos otros estilos

weights = [light,light_medium,medium,heavy,medium_heavy]
labels = ['light','light_medium','medium','heavy','medium_heavy']

explode = [.4,.1,.1,.4,.1]  # Deja dos menos hacia afuera

plt.title("Weight Distrubution of FIFA Players (in kg)")

plt.pie(weights,labels=labels,autopct="%.2f %%",explode=explode)

plt.show()
```

```python
barcelona = fifa.loc[fifa.nationality=='Argentina']['overall_rating']
madrid = fifa.loc[fifa.nationality=='France']['overall_rating']
revs = fifa.loc[fifa.nationality=='Italy']['overall_rating']

labels = ['FC Barcelona','Real Madrid','New England Revolution']

plt.style.use('default')

plt.figure(figsize=(7,10))

boxes = plt.boxplot([barcelona, madrid, revs],labels=labels, patch_artist=True)  # Configure esto verdadero debido a la espalda para configurar Facecolor    

for box in boxes['boxes']:  #          
    # Establecer el color del límite de la caja.
    box.set(color='#4286f4',linewidth=2)
    # Configurar caja de color interno
    box.set(facecolor='#e0e0e0')

plt.title("Professional Soccer Team Comparison")
plt.ylabel("FIFA Overall Rating")

plt.show()
```
